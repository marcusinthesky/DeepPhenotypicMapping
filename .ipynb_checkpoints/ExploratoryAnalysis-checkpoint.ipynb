```python
#Import data
path = os.path.join('.','Data','LabelledData.csv')
data = pd.read_csv(path, index_col='Unnamed: 0')
data = data.loc[data.Machine==0,:].dropna()
```

```python
#Relabel
path = os.path.join('.','Data','Names.csv')
names = pd.read_csv(path, sep=';', index_col='X')
names.index = pd.Series(names.index).apply(lambda x: x[3:])
data = data.replace({'EXPERIMENT':names.to_dict()['Name']})
```
